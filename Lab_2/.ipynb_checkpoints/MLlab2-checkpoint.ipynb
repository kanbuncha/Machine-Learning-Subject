```python
stock_MSFT = pd.read_csv('MSFT.csv')
stock_MSFT.index = pd.to_datetime(stock_MSFT.Date)
stock_MSFT
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
stock_MSFT['Close_norm'] = scaler.fit_transform(np.array(stock_MSFT['Close']).reshape(-1, 1))
```

```python
stock_MSFT_Close = pd.DataFrame(stock_MSFT['Close_norm'])
```

```python
Next_N_day = 30
GT = stock_MSFT_Close.iloc[Next_N_day :]
```

```python
stock_MSFT_train_valid = stock_MSFT_Close.iloc[:-Next_N_day]
stock_MSFT_train_valid = stock_MSFT_train_valid.assign(GT=GT.values)

stock_MSFT_test = stock_MSFT_Close.iloc[-Next_N_day:]

stock_MSFT_train_valid
```

```python
# เตรียมข้อมูล train, validate
from sklearn.model_selection import train_test_split

X_train, X_valid, y_train, y_valid = train_test_split(
    stock_MSFT_train_valid.Close_norm,
     stock_MSFT_train_valid.GT,
    test_size=0.33, random_state=42)
```
